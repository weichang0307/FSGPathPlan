```python
from tqdm import tqdm
from track_generate import FSTrackGenerator
import pandas as pd
```

```python
gen = FSTrackGenerator(
    track_width=4,       # 3m minimum, usually 3.5-4m
    cone_spacing=3,      # Max 5m allowed
    num_control_points=20, 
    randomness=0.6,
    min_turn_radius=3.5    # Enforce 3.5m radius safety margin (Rule is 3m)
)
for i in tqdm(range(20), desc="Generating Track"):
    gen.generate()
    gen.save_to_csv(f"tracks/fs_track_{i}.csv")
```

```python
for i in range(20):
    gen = FSTrackGenerator()
    df = pd.read_csv(f"tracks/fs_track_{i}.csv")
    gen.load_from_dataframe(df)
    gen.plot()
```
